```python
'''
============================================================
DEG count visualization across conditions (heatmaps)

Purpose:
This script visualizes the number of differentially expressed
genes (DEGs) per cell type across experimental conditions
(CCI and SNI) using heatmaps.

Input:
- Excel file containing DEGs from Seurat FindMarkers output

Output:
- Two heatmaps (CCI and SNI) showing DEG counts per cell type

Additional Notes:
- Gene counts are aggregated per (cell_type x condition)
- A shared color scale is used for comparability across plots
============================================================
'''
```

```python
# 0. Setup ####
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# 1. Load input data ####
file_path = ("path/tocondition_vs_naive_DEG_by_celltype.xlsx")
df = pd.read_excel(file_path)


# 2. Aggregate DEG counts per cell type and condition ####
deg_counts = (
    df.groupby(["cell_type", "condition"])["gene"]
    .nunique()
    .reset_index()
)

deg_matrix = deg_counts.pivot(
    index="cell_type",
    columns="condition",
    values="gene"
).fillna(0)


# 3. Define condition groupings ####
cci_order = ["cci_acute", "cci_mid", "cci_chronic"]
sni_order = ["sni_acute", "sni_mid", "sni_chronic"]

deg_matrix_cci = deg_matrix.reindex(columns=cci_order)
deg_matrix_sni = deg_matrix.reindex(columns=sni_order)


# 4. Define cell type order ####
desired_order = [
    "Neurons",
    "Endothelial",
    "Pericytes",
    "Fibroblasts",
    "mSchwann",
    "Sat Glia & nSchwann",
    "RBCs",
    "Mono, Mac, & DC",
    "B Cells",
    "T Cells",
    "Neutrophils"
]

deg_matrix_cci = deg_matrix_cci.reindex(desired_order)
deg_matrix_sni = deg_matrix_sni.reindex(desired_order)


# 5. Define shared color scale ####
vmin = 0
vmax = max(
    deg_matrix_cci.max().max(),
    deg_matrix_sni.max().max()
)


# 6. Heatmap: CCI ####
plt.figure(figsize=(8, 6))

sns.heatmap(
    deg_matrix_cci,
    annot=False,
    fmt=".0f",
    cmap="viridis",
    vmin=vmin,
    vmax=vmax
)

plt.title("CCI DEG Counts")
plt.ylabel("Cell Type")
plt.xlabel("Condition")

plt.savefig(
    "E:/lanah/internal_datasets/da/batch1_3_6_w_additonal_seq/1)allcells/"
    "a16_deg_(condition_vs_naive)_spli_by_celltype_plot/CCI_DEG_heatmap.pdf",
    bbox_inches="tight"
)

plt.show()


# 7. Heatmap: SNI ####
plt.figure(figsize=(8, 6))

sns.heatmap(
    deg_matrix_sni,
    annot=False,
    fmt=".0f",
    cmap="viridis",
    vmin=vmin,
    vmax=vmax
)

plt.title("SNI DEG Counts")
plt.ylabel("Cell Type")
plt.xlabel("Condition")

plt.savefig(
    "E:/lanah/internal_datasets/da/batch1_3_6_w_additonal_seq/1)allcells/"
    "a16_deg_(condition_vs_naive)_spli_by_celltype_plot/SNI_DEG_heatmap.pdf",
    bbox_inches="tight"
)

plt.show()
```
